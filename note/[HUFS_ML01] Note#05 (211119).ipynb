### 2. 프로젝트 방향성에 대한 대책 논의

* 각 조원들의 의견을 공유한 결과, 현재의 데이터셋만으로는 유의미한 분석 결과를 낼 수 없다는 결론에 도달하였고, 이에 따라 코로나 19 상황과 게임 이용 추이에 대해 접근하는 방향에서 크게 벗어나지 않은 새로운 데이터셋에 대한 논의를 진행하였다.
* 조사해온 논문과 다양한 기사에서는 코로나 19 상황에서 게임 이용자 수가 늘었음을 나타내고 있었지만, 현재까지의 분석 상황에서 이러한 결과가 나오지 않은 것은 PC 게임 위주의 데이터셋만을 사용했기 때문이라는 이유로 의견이 모여졌다.
* 이에 특정 플랫폼의 데이터를 활용하기 보다는, 게임 업계의 전반적인 성장세와 하락세를 나타내는 지표인 주식 정보를 활용하여 분석을 하기로 하였다.
* 세부적인 분석 방향성으로는 코로나로 인해서 야외활동이 제한되었기 때문에 게임이 대체재로 작용했는지, 또한 그 중에서도 어느 종류의 게임들이 뚜렷한 성장세를 보였는지를 알아보기로 하였다.
* 또한, 데이터셋을 새로 찾아서 진행하게 되었기 때문에 11월 21일 일요일 오후 9시에에 회의를 한 번 더 진행하기로 하였다.

### 3. 역할 분담

* 문은정 조원은 게임사, 혹은 게임과 관련된 기업들을 조사하여 해당 기업들의 Ticker 코드를 조사하고, 가능하다면 추가적으로 해당 기업에서 주력으로 다루는 게임 관련 컨텐츠가 무엇인지에 대해 조사하기로 하였다.
* 김건호 조원은 주식 데이터를 받아오는 최적의 방안(Python yfinance 패키지, 금융 페이지 크롤링 등)을 조사한 뒤 대표적인 게임 회사의 주식 데이터를 가져오는 역할을 진행하기로 하였다.
* 김한나 조원과 정원창 조원은 주식 데이터와 코로나 데이터셋을 함께 엮어서 분석하기에 어떠한 기계학습 방법론이 적절한지에 대해 조사를 하기로 하였다.
