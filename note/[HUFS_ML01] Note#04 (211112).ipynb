### 2. EDA 결과에 대한 의견 공유

* 데이터셋 별로 진행된 시각화 자료의 월별 수치를 비교 분석해본 결과, 코로나 발생 직후는 게임 이용자 수가 확연히 증가하였으나 코로나에 대한 인식 변화, 거리두기 단계 조정, 백신 접종 등의 이유로 상승세가 계속해서 유지되지 못한 모습을 발견할 수 있었다.
* 이는 게임과 코로나에 대해 다룬 다양한 논문에서 '코로나로 인해 이전보다 더 많은 사람들이 게임을 즐기게 되었을 것'이라는 내용에 완벽하게는 대응되지 않는 결과로 비춰졌다.
* 이에 대해 단순히 '코로나 확진자 수와 게임 이용자 수는 정비례'라는 접근을 하기 보다는, 두 데이터셋을 함께 연관지어 분석하며 국가별 & 기간별 데이터를 상세히 관찰함과 동시에 그래프만으로는 파악되지 않는 관계를 기계학습 방법론을 통해 분석해보자는 결론에 도달하였다.

### 3. 다음 단계에 대한 역할 분담

* 김건호 조원은 tg_df에 추가할 수 있는 속성들이 무엇인지에 대해 알아본 뒤 tg_df에 대해 다양한 기계학습 모형을 적용해보기로 하였다.
* 정원창 조원과 문은정 조원은 함께 steam 플랫폼의 데이터 이외에 다른 플랫폼 관련 데이터를 찾아본 뒤 분석 대상 게임을 늘릴 수 있는 방안과 tg_df에 추가할 수 있는 속성들을 조사하기로 하였다.
* 김한나 조원은 이전의 탐색적 데이터 분석 결과를 통해 스팀 게임을 가장 많이 이용하는 미국에서의 확진자 수 그래프와 소셜 게임의 이용자 그래프 수가 어느 정도 비슷한 양상을 띈 다는 것을 확인하였고, 이에 미국 확진자 수와 소셜 게임 이용자 수의 관계를 더 구체적인 수치로 알아보면서 그 밖에 국가별 Steam 게임 이용 비율 대로 국가별 확진자 수를 가중 시킨 값이 게임 이용량과 관계가 있는지를 살펴보기로 하였다.
