### 1. 정기 회의 관련

* 각 조원들의 시간표를 취합하여 정기 회의 날짜를 매 주 금요일 오후 4시~5시 30분으로 결정하였고, 시험 기간과 졸업 논문 제출을 고려하여 2021.10.29 금요일에 진행하였다.
* 팀 프로젝트를 진행하면서 논의가 더 필요한 부분이 생기거나 시간 내에 정기 회의를 끝내지 못하는 경우, 유동적으로 추가 회의를 진행하기로 하였다.



---



### 2. 프로젝트 주제 조사

* 일주일 동안 팀 프로젝트의 주제에 대한 아이디어를 각각 2~3개씩 조사하기로 하였다.
* 주제 조사 내용을 문서화 하여 사전에 공유를 한 뒤, 다음 정기 회의에 의논을 통해 주제를 확정하기로 하였다.
* 데이터 셋 제공 사이트 : [kaggle](https://www.kaggle.com/), [UCI](https://archive.ics.uci.edu/ml/index.php), [빅데이터센터](https://kbig.kr/portal/kbig/datacube/dataset/info)

### 4. 역할 분담

| 역할           | 내용                                                         | 담당           |
| -------------- | ------------------------------------------------------------ | -------------- |
| 발표           | 발표 PPT 최종 검수, 발표 진행                                | 김한나, 정원창 |
| 최종 코드 관리 | 최종 제출 코드의 주석 및 작성 스타일 통일,<br />깃헙 이력 관리 및 충돌 방지 | 김건호         |
| 데이터셋 관리  | 전처리 방법 및 적절한 모형 아이디어 제시<br />데이터셋의 칼럼 정보 심층 분석 | 문은정         |

* 비대면으로 진행되는 팀 프로젝트의 특성 상 자료 조사부터 코드 작성, 발표에 이르기까지의 전 과정을 유기적으로 함께 진행할 수 없기 때문에, 진행 과정 별로 주도적으로 관리를 맡는 역할을 선정하였다. 
* 이에 따라 팀 프로젝트에 제출해야하는 3가지 파일(발표 파일, 분석 코드, 분석에 사용된 데이터)에 맞게 역할을 나눈 뒤 수요 조사를 통해 각각의 역할로 분담을 하였다.
* 모든 조원은 진행 과정에 따라 함께 작업을 진행하고 각 역할별로 조원들이 역할에 맞는 추가 검수를 진행하게 된다.
