```python
import datetime

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
c1 = pd.read_csv("Corona2020.csv", encoding = 'euc-kr')
c2 = pd.read_csv("Corona2021.csv", encoding = 'euc-kr')
```

```python
df = pd.merge(c1, c2, on = ['국가별', '항목', '단위'])
```

```python
list(df['국가별'])
```

```python
c = pd.DataFrame()
countrylst = ['미국','중국', '러시아', '브라질', '독일', '캐나다', '프랑스', '영국', '폴란드', '터키', '필리핀', '우크라이나', '한국', '호주', '인도네시아']
for x in countrylst:
    c = c.append(df[(df['국가별'] == x)])
```

# 각 날짜별 확진자 수 추이

```python
ill['date'] = pd.to_datetime(ill['date'], format='%Y. %m. %d')
df = ill
```

```python
df = df.diff()
df['date'] = ill['date']
df = df[1:]
```

```python
column_list = list(df)
column_list.remove("date")
```

# 오류 처리, 0보다 작은 확진자 수는 0으로 처리

```python
for i in column_list:
    a = i
    for j in range(1,len(df[a])):
        if df[a][j] < 0:
            df[a][j] = 0
```

```python
df["sum"] = df[column_list].sum(axis=1)
```

```python
sns.set(rc={'figure.figsize':(20,10)})
```

```python
ax = sns.lineplot(x="date", y="sum",data=df)
ax.set(xlabel='Date', ylabel='Cases', title='Cases from 15 countries')
```

```python
df['month'] = df['date'].apply(lambda t: t.month)
df['year'] = df['date'].apply(lambda t: t.year)
```

```python
df
```

```python
df2 = df.groupby(['year', 'month']).mean()
df2
```

```python
df2 = df2.drop('sum', axis = 1)
```

```python
sns.set(rc={'figure.figsize':(20,12)})
```
