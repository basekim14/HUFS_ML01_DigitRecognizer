```python
#weekend 추가
for i in range(len(traffic.date)):
    traffic.date[i] = datetime.datetime.strptime(traffic.date[i], '%Y-%m-%d')
    whatday = traffic.date[i].weekday()
    if (whatday == 5) or (whatday == 6):
        traffic.weekend[i] = 1
traffic
```
