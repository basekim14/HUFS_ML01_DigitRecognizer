### 2. 프로젝트 주제 선정

* 주제별 인사이트를 의논한 뒤 최종 투표를 거친 결과 '언택트 시대와 게임 플레이 시간의 추이' 주제가 팀 프로젝트의 주제로 선정이 되었다.

### 3. 데이터셋에 대한 EDA 진행

* '언택트 시대와 게임 플레이 시간의 추이' 주제의 세부적인 분석 방향성과 적합한 모델을 찾는 것으로 해당 주제에 대한 유의미한 결론을 이끌어낼 수 있게 하기 위해 해당 주제에 사용된 데이터셋에 대한 EDA를 진행하기로 하였다.
  * 데이터셋 링크 : [월별 스팀 게임 이용자 데이터](reference/AllSteamData.csv), [코로나 19 발생 현황(2020)](reference/101_DT_COVID19_004_20211001150023.csv), [코로나 19 발생 현황(2021)](reference/101_DT_COVID19_004_20211001150440.csv) 
* 먼저, '월별 스팀 게임 이용자 데이터'에 대한 EDA는 김건호 조원, 정원창 조원이 맡게 되었으며 이용자 데이터 외에 게임에 관한 데이터를 추가적으로 스크래핑 한 뒤 분석에 사용할 항목들과 독립 변수들에 대해 탐색을 하기로 하였다.
* 그리고 '코로나 19 발생 현황'에 대한 EDA는 문은정 조원, 김한나 조원이 맡게 되었으며 전세계 국가의 일별 · 월별로 누적확진자와 신규 확진자를 살펴보면서 전세계 코로나 확진자 수의 추이를 파악해본 뒤, Steam의 주요 이용 국가 15개국의 확진자 수의 추이를 파악하면서 Steam 게임 사용자수와 코로나 확진자 수 간의 관계를 중점적으로 파악하고자 한다.
  * United States, China, Russia, Brazil, Germany, Canada, France, United Kingdom, Poland, Turkey, Philippines, Ukraine, Korea, Australia, Indonesia
